## Hard margin SVM

模型的决策函数：

$$
    f(x) = sign(w^\text T x + b)
$$

最大间隔分类器：

$$
    \max \text{margin} (w, b) \\
    \mathrm{s.t.} \  y_i(w^\text T x_i + b) > 0
$$

间隔可以表示为$\text{margin}(w, b)$:

\begin{align}
    \text{margin}(w, b) & = \min \text{distance}(w, b, x_i) \\
    & = \min \frac{1}{\Vert w \Vert} \lvert w^\mathrm T x_i + b \rvert
\end{align}

$\therefore$ 最大间隔为：

\begin{align}
    & \max_\limits{w, b} \min_\limits{x_i} \frac{1}{\Vert w \Vert} \lvert w^\mathrm T x_i + b \rvert \\
    \Longrightarrow & \max_\limits{w, b} \frac{1}{\Vert w \Vert} \min_\limits{x_i} y_i(w^\mathrm T x_i + b)
\end{align}

令 $y_i(w^\mathrm T x_i + b) = \gamma$，即 $\exists \gamma > 0$，使 $\mathrm{s.t.} \min y_i(w^\mathrm T x_i + b) = \gamma$

令 $\gamma = 1$，即固定 $w$ 范数
